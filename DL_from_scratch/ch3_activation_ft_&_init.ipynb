4x3 짜리 weight & 2x4 짜리 weight

np 일때가 np.random.randn()

torch 일때는 torch.randn() -> 정규분포에서 샘플링

```python
import torch
w1 = torch.randn(3,4)
w2 = torch.randn(4,2)

print(w1)
```

초기화 방법 3가지 기억하지? xaiver, he, lacun

```python
import torch.nn as nn
# Xavier

w1 = torch.empty(3,4)
nn.init.xavier_normal_(w1)
nn.init.xavier_normal_(w2)

print(w1)
```

위에 정규분포 시각화 한번 해볼까?

```python
# 1. 우선 정규분포를 그린다
# 어케그릴건데..? 식 넣고 0부터 쭉 샘플링 해볼까?

import math
def normal_distribution(x,mu=0,sigma=1):
    return (1/(math.sqrt(2*math.pi))*sigma)*torch.exp(-0.5*(x-mu)**2/sigma**2)
```

ndarray -> tensor : torch.tensor()

```python
x = np.arange(-10,10,0.1)
x = torch.tensor(x)
y = normal_distribution(x)
plt.plot(x,y)
plt.show()
```

```python
x1 = np.arange(-10,10,0.1)
x1 = torch.tensor(x1)

#2. 나온 weight를 펼쳐서 점을 찍어본다.

w1 = w1.flatten()
y1 = normal_distribution(x1)
y2 = normal_distribution(w1)
plt.plot(x1,y1)
plt.scatter(w1,y2,color="red", label="Sampled Weights")
plt.show()
```
